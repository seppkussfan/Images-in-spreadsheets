```python
# Get picture and resize it to 256 x 144

from PIL import Image

im = Image.open(r"picture_3.jpg")

im_resized = im.resize((640, 480))
im2 = im_resized.resize((80, 240))
pixels = list((im2.getdata()))
```

```python
# creating a list to identify all the names of the 256 columns

column_list = []
column = "A"
for i in range(80):
    column_list.append(column)
    
    if column[-1] == 'Z':
        if len(column) == 1:
            column = "AA"
        else:
            column = chr(ord(column[0]) + 1) + 'A'
    else:
        if len(column) == 1:
            column = chr(ord(column) + 1)
        else:
            column = column[0] + chr(ord(column[-1]) + 1)

# print(len(column_list))
```

```python
# Writing the color in the pixels list to the spreadsheet.

from openpyxl import Workbook
from openpyxl.styles import PatternFill

wb = Workbook()
sheet = wb.active

row = 0
for i in range(len(pixels)):
    if i % 80 == 0:
        row = row + 1
    
    # obtaining r, g, b for each cell
    column = column_list[i % 80]
    pixel = pixels[i]
    r = hex(pixel[0])[2:]
    g = hex(pixel[1])[2:]
    b = hex(pixel[2])[2:]
    
    # zero padding each section of the aRBG hex
    r = f'{r:0>2}'
    g = f'{g:0>2}'
    b = f'{b:0>2}'
    color_code = "00" + r + g + b 
    color_fill = PatternFill(start_color=color_code, end_color=color_code, fill_type="solid")
    
    # actual filling into cell
    cell = column + str(row)
    sheet[cell].fill = color_fill
        
wb.save(r"aa.xlsx")
```
